```python
def Ratio(environnement):
    df=pd.DataFrame(environnement)
    equipe=df[df["teammate"]==True].shape[0]
    adverse=df[df["teammate"]==False].shape[0]
    return equipe/adverse
```

```python
def listratio(listenvironnement):
    result=[]
    for i in range(listenvironnement.shape[0]):
        df=pd.DataFrame(listenvironnement.iloc[i])
        equipe=df[df["teammate"]==True].shape[0]
        adverse=df[df["teammate"]==False].shape[0]
        result+=[equipe/adverse]
    return result
                    
listratio(df["environnement"])
```
